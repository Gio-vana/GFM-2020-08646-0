### Rede Neural Artificial

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score, f1_score, precision_score, accuracy_score
```

```python
indicadores_mglu3_treinamento.shape
```

```python
model = Sequential()
model.add(Dense(16, activation='relu', input_dim= 4))
model.add(Dense(16, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(indicadores_mglu3_treinamento, target_mglu3_treinamento, epochs=10, batch_size=128, validation_data=(indicadores_mglu3_teste, target_mglu3_teste))
```

```python
model.summary()
```

```python
yajustado = model.predict_classes(indicadores_mglu3_treinamento)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import recall_score, f1_score, precision_score
```

```python
matriz_confusão = confusion_matrix(yajustado,target_mglu3_treinamento)
print(f'Falsos positivo - fp: {matriz_confusão[0,1]}')
print(f'Verdadeiros positivo - tp: {matriz_confusão[1,1]}')
print(f'Falsos negativo - fn: {matriz_confusão[1,0]}')
print(f'Verdadeiros negativo  - tn: {matriz_confusão[0,0]}')
print(f'Recall: {recall_score(target_mglu3_treinamento, yajustado)}')
print(f'F1-score: {f1_score(target_mglu3_treinamento, yajustado)}')
print(f'Precisão: {precision_score(target_mglu3_treinamento, yajustado)}')
print(f'Acurácia: {accuracy_score(target_mglu3_treinamento, yajustado)}')
```

```python
y_predito = model.predict_classes(indicadores_mglu3_teste)
```

```python
y_predito
```

```python
matriz_confusão = confusion_matrix(y_predito,target_mglu3_teste)
print(f'Falsos positivo - fp: {matriz_confusão[0,1]}')
print(f'Verdadeiros positivo - tp: {matriz_confusão[1,1]}')
print(f'Falsos negativo - fn: {matriz_confusão[1,0]}')
print(f'Verdadeiros negativo  - tn: {matriz_confusão[0,0]}')
print(f'Recall: {recall_score(target_mglu3_teste, y_predito)}')
print(f'F1-score: {f1_score(target_mglu3_teste, y_predito)}')
print(f'Precisão: {precision_score(target_mglu3_teste, y_predito)}')
print(f'Acurácia: {accuracy_score(target_mglu3_teste, y_predito)}')
```

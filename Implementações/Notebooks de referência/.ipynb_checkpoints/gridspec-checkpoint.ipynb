# Exemplo de múltiplos plots

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
```

## Vamos construir algumas funções

```python
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1*y2
```

## Definindo o grid (duas funções, lado a lado)

```python
fig = plt.figure(figsize=[10, 4])
grid = gridspec.GridSpec(1, 2)  # uma linha, duas colunas  [ FIG1   FIG2 ]

ax = fig.add_subplot(grid[0, 0])
ax.plot(x, y1)
ax.set_title('$y_1=\sin(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])

ax = fig.add_subplot(grid[0, 1])
ax.plot(x, y2)
ax.set_title('$y_2 = \cos(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])
```

## Duas funções, vertical

```python
fig = plt.figure(figsize=[4, 10])
grid = gridspec.GridSpec(2, 1)  # duas linha, uma coluna

ax = fig.add_subplot(grid[0, 0])
ax.plot(x, y1)
ax.set_title('$y_1=\sin(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])

ax = fig.add_subplot(grid[1, 0])
ax.plot(x, y2)
ax.set_title('$y_2 = \cos(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])
```

## Os grids não precisam ser igualmente espaçados

```python
fig = plt.figure(figsize=[10, 10])
grid = gridspec.GridSpec(2, 2)  # interprete ;-)

ax = fig.add_subplot(grid[0, 0])
ax.plot(x, y1)
ax.set_title('$y_1=\sin(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])

ax = fig.add_subplot(grid[0, 1])
ax.plot(x, y2)
ax.set_title('$y_2 = \cos(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-1, 0, 1])

ax = fig.add_subplot(grid[1, :])
ax.plot(x, y3)
ax.set_title('$y_3 = \sin(x)\,\cos(x)$')
ax.set_xticks([-2*np.pi, 0, 2*np.pi])
ax.set_xticklabels(['$-2\pi$', '$0$', '$2\pi$'])
ax.set_yticks([-0.5, 0, 0.5])
ax.set_yticklabels(['$-\\frac{1}{2}$', '$0$', '$\\frac{1}{2}$'])
```
